```python
import pandas as pd
import numpy as np
import requests
```

```python
# making dataframes from StreamingHistory files
df_stream0 = pd.read_json('StreamingHistory0.json')
df_stream1 = pd.read_json('streamingHistory1.json')

# merge streaming dataframes
df_stream = pd.concat([df_stream0, df_stream1])

# combining the fields 'artistName' and 'trackName'
df_stream['albumName'] = df_stream['artistName'] + ":" + df_stream['trackName']

df_stream.head()
```

```python
import json

# Assuming you have your JSON data stored in a file named "YourLibrary.json"
with open("YourLibrary.json") as file:
    json_data = json.load(file)
    tracks = json_data['tracks']

    # Extracting and printing all the tracks
    print("All Tracks:")
    for track in tracks:
        print("Track:", track['track'])
        print("Artist:", track['artist'])
        print("Album:", track['album'])
        print("URI:", track['uri'])
        print()
```

```python
# Group by artistName and calculate the sum of minutes and streams
df_top_artists = df_stream.groupby('artistName').agg({'msPlayed': 'sum', 'trackName': 'count'}).reset_index()

# Rename columns
df_top_artists.rename(columns={'msPlayed': 'Total Minutes', 'trackName': 'Total Streams'}, inplace=True)

# Sort by Total Minutes in descending order
df_top_artists.sort_values('Total Minutes', ascending=False, inplace=True)

# Display the dataframe
print("Top Artists:")
print(df_top_artists.head())  
```

```python
# Group by trackName and calculate the sum of minutes and streams
df_top_songs = df_stream.groupby('trackName').agg({'msPlayed': 'sum', 'artistName': 'count'}).reset_index()

# Rename columns
df_top_songs.rename(columns={'msPlayed': 'Total Minutes', 'artistName': 'Total Streams'}, inplace=True)

# Sort by Total Minutes in descending order
df_top_songs.sort_values('Total Minutes', ascending=False, inplace=True)

# Display the dataframe
print("Top Songs:")
print(df_top_songs.head())
```

```python
# calculate total minutes per year
total_minutes_year = df_stream['msPlayed'].sum() / (1000 * 60)
print("Total Minutes on Spotify for the Year:", total_minutes_year)
```

```python
# calculate total minutes per month
df_stream['endTime'] = pd.to_datetime(df_stream['endTime'])
df_stream['month'] = df_stream['endTime'].dt.month

total_minutes_month = df_stream.groupby('month').agg(total_minutes=('msPlayed', 'sum')) / (1000 * 60)
print("Total Minutes on Spotify per Month:")
print(total_minutes_month)
```

```python
# calculate total minutes per day
df_stream['day'] = df_stream['endTime'].dt.date

total_minutes_day = df_stream.groupby('day').agg(total_minutes=('msPlayed', 'sum')) / (1000 * 60)
print("Total Minutes on Spotify per Day:")
print(total_minutes_day)
```

```python
# calculate total minutes per week
df_stream['week'] = df_stream['endTime'].dt.week

total_minutes_week = df_stream.groupby('week').agg(total_minutes=('msPlayed', 'sum')) / (1000 * 60)
print("Total Minutes on Spotify per Week:")
print(total_minutes_week)
```

```python
# Extract unique genres from track names
df_stream['genre'] = df_stream['trackName'].str.split('-', expand=True)[0].str.strip()

unique_genres = df_stream['genre'].nunique()

print("Unique Genres:", unique_genres)
```

```python
unique_tracks = df_stream['trackName'].nunique()
unique_albums = df_stream['albumName'].nunique()
unique_genres = df_stream['genre'].nunique()
unique_artists = df_stream['artistName'].nunique()

print("Unique Tracks:", unique_tracks)
print("Unique Albums:", unique_albums)
print("Unique Genres:", unique_genres)
print("Unique Artists:", unique_artists) 
```
